# 시도, 시군구, 읍면동으로 나누기

```python
concat_df["t1"]=concat_df['시군구'].str.split(" ").str[0]
def getSiGunGu(value):
    sido = value.split(" ")[0]
    val = value.split(" ")[1]
    if len(val)>=5:
        return val[:2]+"시 "+val[2:]
    else:
        if "세종" in sido:
            return "전체"
        else:
            return val
def getEupMeonDong(value):
    sido = value.split(" ")[0]
    
    if "세종" in sido:
        return value.split(" ")[1]
    else:
        return value.split(" ")[2]
    
```

```python
concat_df["t2"] = concat_df['시군구'].apply(getSiGunGu)
```

```python
concat_df["t3"] = concat_df['시군구'].apply(getEupMeonDong)
```

```python
concat_df["도로명"] = concat_df['도로명'].str.strip()
concat_df["t1"] = concat_df['t1'].str.strip()
concat_df["t2"] = concat_df['t2'].str.strip()
concat_df["t3"] = concat_df['t3'].str.strip()
```

# 위도 경도 때려넣기

```python
import requests
import json

# 카카오 API 써서 주소가 덜 작성된거는 위도 경도로 새 주소로 업뎃해쥼
headers = {
"Authorization":"KakaoAK 66447c8b7057c042f65ac177c53ea385"
}
```

```python
address = {}
def getLatLong(addr):
    if addr in address:
        return address[addr]
    else:
        params = {'query': addr}
        response = requests.get('https://dapi.kakao.com/v2/local/search/address.json', headers=headers, params=params)
        count = response.json()['meta']['total_count']
        if count!=0:
            res = response.json()
            try:
                lat =res['documents'][0]['address']['y']
                lon = res['documents'][0]['address']['x']
                address[addr] = lon+"/"+lat
                return lon+"/"+lat
            except:
                print(res)
                return "ERROR"
        else :
            return "NO"

concat_df['latlon'] = concat_df['도로명'].apply(getLatLong)
```
